```python
# datasets

df = pd.DataFrame(data = load_wine().data, columns = load_wine().feature_names)
df.head()
```

```python
# combine data with label

df["results"] = load_wine().target

df.head()
```

```python
# datasets separations
```

```python
X = df.drop("results",axis = 1).values
y = df["results"].values
```

```python
X.shape, y.shape
```

```python
y.shape
```

```python
# train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state= 35)
```

```python
X_train.shape, y_test.shape
```

```python
# minmax scaler

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# data to scaler

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# testing min max 

X_train_scaled.min(), X_train_scaled.max()
```

```python
# convert results to categorical

from tensorflow.keras.utils import to_categorical
```

```python
y_train_cat = to_categorical(y_train) 
y_test_cat = to_categorical(y_test)
```

```python
# model

from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.models import Sequential
```

```python
X.shape, y.shape
```

```python
model = Sequential([
    Dense(34, activation = "relu"),
    Dense(89, activation = "relu"),
    Dense(3, activation = "softmax"),
])
```

```python
model.compile(
    optimizer= "adam",
    loss = "categorical_crossentropy",
    metrics = ["accuracy"]
)
```

```python
# early stopping in keras

from tensorflow.keras.callbacks import EarlyStopping
```

```python
early = EarlyStopping(monitor = "val_loss",patience= 25,mode = "min", verbose = 1)
```

```python
# fitting the model
```

```python
Model = model.fit(X_train_scaled ,y_train_cat , epochs = 300, validation_data= (X_test_scaled, y_test_cat),callbacks= [early])
```

```python
pd.DataFrame(Model.history).plot()
```

```python
# evaluation
```

```python
from sklearn.metrics import classification_report, confusion_matrix

```

```python
# let's predicts the data

y_pred = model.predict_classes(X_test_scaled)
```

```python
classification_report(y_pred, y_test)
```

```python
confusion_matrix(y_pred, y_test)
```

```python
# let use different optimizer and hidden layer
```

```python

Model_train = {}

for optimizers in ["adam","rmsprop"]:
    for hidden_layer in [100, 250, 300]:
        
        Model_Historys = []
        Model_confusion_matrix = []

        
        # model layers
        model = Sequential([
            Dense(35, activation = "relu"),
            Dense(hidden_layer+50, activation = "relu"),
            Dense(hidden_layer, activation = "relu"),
            Dense(3, activation = "softmax")
        ])
        
        model.compile(
                    optimizer= optimizers,
                    loss = "categorical_crossentropy",
                    metrics = ["accuracy"]
        )
        Model = model.fit(X_train_scaled, y_train_cat, epochs = 300, validation_data= (X_test_scaled, y_test_cat), callbacks= [early])
        # getting the loss
        Model_Historys.append(Model.history)
        # testing the test data
        y_pred = model.predict_classes(X_test_scaled)
        Model_confusion_matrix.append(confusion_matrix(y_pred, y_test))
        Model_train[optimizers+""+str(hidden_layer)] = [Model_Historys, Model_confusion_matrix]
        
```
