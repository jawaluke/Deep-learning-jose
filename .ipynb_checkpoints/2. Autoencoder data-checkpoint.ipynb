```python
# model layers
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
#  3---> 2---> 3
```

```python
encoder = Sequential([
    Dense(2, activation = "relu", input_shape = [15]),
])
```

```python
decoder = Sequential([
    Dense(15, activation = "relu", input_shape = [2])
])
```

```python
autoencoder = Sequential([encoder, decoder])
```

```python
autoencoder.compile(
    optimizer="adam",
    loss = "mse"
)
```

```python
# scaling data

from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled.min(), X_scaled.max()
```

```python
#fitting the model
```

```python
autoencoder.fit(X_scaled, X_scaled, epochs = 9)
```

```python
# only encoded section
```

```python
encoded_2dim = encoder.predict(X_scaled)
```
